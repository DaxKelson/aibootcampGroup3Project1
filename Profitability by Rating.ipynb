```python
import pandas as pd
import os
```

```python
start_folder = 'archive/Data'

# Create a dataframe to store the merged data
imdb_all_years_df = pd.DataFrame()

# Loop through each folder
for year in os.listdir(start_folder):
    year_df = pd.read_csv(os.path.join(start_folder, year, 'merged_movies_data_' + year + '.csv'))
    imdb_all_years_df = pd.concat([imdb_all_years_df, year_df])
imdb_all_years_df
```

```python
imdb_dropped_df = imdb_all_years_df.dropna(how='any')
imdb_dropped_df
```

```python
imdb_SKorea_df = imdb_dropped_df.query("countries_origin.str.contains('South Korea')")
imdb_SKorea_df
```

```python
imdb_Sweden_df = imdb_dropped_df.query("countries_origin.str.contains('Sweden')")
imdb_Sweden_df
```

```python
#Removing all Country Origins not containing US and/or Canada/UK
imdb_US_UK_CA_others_df = imdb_dropped_df[imdb_dropped_df['countries_origin'].apply(lambda x: 'United States' in x or 'United Kingdom' in x or 'Canada' in x)]
imdb_US_UK_CA_others_df
```
