```python
# Define x and y data
x2 = imdb_df['ProfitWorldWide']
y = imdb_df['Rating']

# Scale size based on 'US Canada Profit'
sizes = (x2 / max(x2) * 500 + 30)  # Adjust the scaling factor as needed

# Create the scatter plot
plt.figure(figsize=(20, 6), facecolor='#f6f6f1')
plt.gca().set_facecolor('#f6f6f1')
plt.scatter(x2, y, s=sizes, alpha=0.25, c='#496649', edgecolors='#303030')

# Labels and Title
plt.xlabel("Profit in Millions", fontsize=18, c='#303030')
plt.ylabel("IMDB Rating", fontsize=18, c='#303030')
plt.title("Correlation between Profit and Rating", fontsize=20, c='#303030')

# Display plot
plt.show()
```
