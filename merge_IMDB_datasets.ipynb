```python
import pandas as pd
import os
```

```python
start_folder = 'archive/Data'

# Create a dataframe to store the merged data
imdb_all_years_df = pd.DataFrame()

# Loop through each folder
for year in os.listdir(start_folder):
    year_df = pd.read_csv(os.path.join(start_folder, year, 'merged_movies_data_' + year + '.csv'))
    imdb_all_years_df = pd.concat([imdb_all_years_df, year_df])
imdb_all_years_df
```

```python
# Export the data to a csv file
imdb_all_years_df.to_csv('imdb_all_years.csv', index=False)
```
